##Nasa POWER

```r
library("tidyverse")
library("nasapower")
```

```r
dates_ = c("1981-01-01","2023-03-31")
install.packages("nasapower", quiet=TRUE)
# Tucuman La Madrid: Longitude -65.2454 Latitude -27.6551 (para googlemaps hay que invertirle el orden, primero Lat y luego Long)
daily_ag <- get_power(community = "ag",
                      # lonlat = c(-65.2454 ,-27.6551), #Lamadrid
                      lonlat = c(-65.764005, -28.454637), #Catamarca
                      pars = c("T2M", "T2MDEW", "T2MWET", "TS", "QV2M", "RH2M",
                               "PRECTOTCORR", "PS", "WS10M", "WS10M_MAX", "WS10M_MIN",
                               "WS10M_RANGE", "WD10M", "WS50M", "WS50M_MAX", "WS50M_MIN",
                               "WS50M_RANGE", "WD50M"),
                      # dates = c("1981-01-01","2023-03-31"),
                      dates = dates_,
                      # dates = c(Cad_inicio, Cad_fin),
                      # Cad_inicio="1981-01-01"
                      # Cad_fin= "2023-03-31"
                      temporal_api = "daily"
)
# daily_ag
daily_ag <- daily_ag %>%
              select(-c(LON, LAT, DOY, YYYYMMDD)) %>%
              rename(MO=MM, DY=DD) %>%
              mutate_at(c('YEAR', 'MO', 'DY'), as.integer)
daily_ag[is.na(daily_ag)] <- -999
write_csv(x = daily_ag, file = "POWER_Point_Daily_20000101_20221212_027d6551S_065d2454W_LST_01.csv")
```
